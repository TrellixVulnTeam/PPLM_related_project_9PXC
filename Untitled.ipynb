```python
import json
from datasets import load_dataset, Dataset
import pandas as pd
```

```python
def load_data(f):
    with open(f+'.jsonl', 'r') as json_file:
        json_list = list(json_file)

    data = []
    for json_str in json_list:
        result = json.loads(json_str)
        if result['relType'] == 'Triple Completion':
            data.append(result)
    with open(f+'.txt','w') as newf:
        for datum in data:
            qa = datum['query'].replace("[MASK]",datum['answer'][0])
            qa = '<|endoftext|>'+datum['question']+'<|endoftext|>'+qa+'<|endoftext|>'
            newf.write(qa+'\n')
```

```python
def load_data_sp(f):
    with open(f+'.jsonl', 'r') as json_file:
        json_list = list(json_file)

    data = []
    for json_str in json_list:
        result = json.loads(json_str)
        if result['relType'] == 'Triple Completion':
            data.append(result)
    with open(f+'_sp.txt','w') as newf:
        for datum in data:
            qa = datum['query'].replace("[MASK]",datum['answer'][0])
            qa = '<|BOS|>'+datum['question']+'<|SEP|>'+qa+'<|endoftext|>'
            newf.write(qa+'\n')
```

```python
def load_data_pd(f):
    with open(f+'.jsonl', 'r') as json_file:
        json_list = list(json_file)

    data = []
    for json_str in json_list:
        result = json.loads(json_str)
        if result['relType'] == 'Triple Completion':
            data.append(result)
    return pd.DataFrame(data)
```

```python
import jsonlines
import json

def jsonl2json(base,inp):
    read_path = base+inp+'.jsonl'
    write_path = base+inp+'.json'
    with jsonlines.open(read_path, "r") as rfd:
        with open(write_path, "w", encoding='utf-8') as wfd:
            for data in rfd:
                json.dump(data, wfd)
```

```python
from collections import defaultdict
def stat_data(f):
    with open(f+'.jsonl', 'r') as json_file:
        json_list = list(json_file)

    stat = defaultdict(int)
    for json_str in json_list:
        result = json.loads(json_str)
        if result['relType'] == 'Triple Completion':
            stat[result['pred']]+=1
    return stat
```
