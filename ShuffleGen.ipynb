```python
import pandas as pd
import json
```

```python
def load_data_pd(f):
    with open(f+'.json', 'r') as json_file:
        json_list = list(json_file)

    data = []
    for json_str in json_list:
        result = json.loads(json_str)
        data.append(result)
    return pd.DataFrame(data)
```

```python
data = load_data_pd('correct_answered')
```

```python
data.head()
```

```python
data['oneanswer'] = data['answer'].apply(lambda x:x[0])
```

```python
pred_set = data.groupby('pred')['oneanswer'].agg(set).reset_index()
```

```python
pred_set
```

```python
pred_set = pred_set[pred_set['oneanswer'].map(len)>=5].reset_index(drop=True)
```

```python
pred_set = pred_set.rename(columns={'oneanswer':'group_ans'})
```

```python
pred_set
```

```python
merged = data.merge(pred_set,on='pred')
```

```python
def shuffleans(origin,ans):
    import random
    ans = list(ans)
    new = random.choice(ans)
    while new == origin:
        new = random.choice(ans)
    return new
```

```python
merged['wrongans'] = merged.apply(lambda row:shuffleans(row['oneanswer'],row['group_ans']), axis=1)
```
